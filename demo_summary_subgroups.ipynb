## Using a 'High-Low List'

Often you won't quite see the default settings produce the shading intensity you'd like for each subgroup.

If that is the case you can provide a list of the subgroups specifying the order. Because this is so important, I have made it an entire section. This will demonstrate it from the command line equivalent and then using the script's main function.

First we need a dataframe that will demonstrate the benefit.

````python
import pandas as pd
sales = [('Jones LLC', 177887, 'yes'),
         ('Jones LLC', 12387, 'yes'),
         ('Jones LLC', 1772287, 'yes'),
         ('Jones LLC', 19897, 'no'),
         ('Jones LLC', 1187, 'no'),
         ('Jones LLC', 1773297, 'maybe'),
         ('Alpha Co', 157987, 'yes'),
         ('Alpha Co', 158981, 'yes'),
         ('Alpha Co', 159983, 'yes'),
         ('Alpha Co', 167987, 'yes'),
         ('Alpha Co', 158117, 'yes'),
         ('Alpha Co', 1999917, 'maybe'),
         ('Alpha Co', 193917, 'maybe'),
         ('Alpha Co', 1933917, 'maybe'),
         ('Alpha Co', 159333, 'no'),
         ('Alpha Co', 256521, 'no'),
         ('Blue Inc', 111947, 'no')]
labels = ['Manufacturer', 'Item', 'In_Stock']
dfh = pd.DataFrame.from_records(sales, columns=labels)
dfh.head()
````

````python
dfh.to_csv('datahl.tsv', sep='\t',index = False)
````

Now to run that without a High-Low list.

````python
%run donut_plot_with_total_summary_and_subgroups_from_dataframe.py datahl.tsv In_Stock Manufacturer 
````

Note that just because of the way the subgroup categories occured in the defining of the dataframe 'maybe' is last in the intensity coloring. Although one would probably want 'no' the least intensely shaded.

This can be fix with the addition of a 'high-low' list.

````python
%run donut_plot_with_total_summary_and_subgroups_from_dataframe.py datahl.tsv In_Stock Manufacturer --hilolist yes,maybe,no
````

Note that now the 'no' subgroup is the lightest shading in all the groups.   
It is important not to place spaces between the commas in the list you provide on the command line.

To do the same with the main function of the script, you provide a Python list as `hilolist`:

````python
from  donut_plot_with_total_summary_and_subgroups_from_dataframe import  donut_plot_with_total_summary_and_subgroups_from_dataframe
h = donut_plot_with_total_summary_and_subgroups_from_dataframe(df=dfh,state4subgroup_col ="In_Stock",grouping_col="Manufacturer", hilolist=["yes", "maybe", "no"]);
````

Note that 'no' subgroup is the latest in all the groups with that. Feel free to remove assigning 'hilolist' and calling the main function to see that change.

## Adjusting the colors used

**Controlling color from the command line.**

The --advance_color option followed by an integer can be added to the call to the script to advance the colors initially used from the sequential color palette generator. This is meant to try to make it easier to customize the output to a color combination that seems pleasing without needing to edit the code in the script. (If you want to specify your own colors, you can edit the list_of_other_good_sequences and run the with -ac 4. [Alternatively you can edit the color_brewer_seq_names and not advance the color generator.])

````python
%run donut_plot_with_total_summary_and_subgroups_from_dataframe.py data.tsv  In_Stock Manufacturer --advance_color 61
````

When fixing the shading order in the subplot on the right above,  a 'high-low' list was added, like so:

````python
%run donut_plot_with_total_summary_and_subgroups_from_dataframe.py datahl.tsv In_Stock Manufacturer --hilolist yes,maybe,no
````

That improved the situation by making the 'no' category the most faintly shaded; however, in the subplot on the left the 'maybe' category is associated with a bright red color that might make more sense as associated with 'no'.

There are a couple of arguments that can be added when calling the script to adjust that.  
One is `--swap_left_colors`:

````python
%run donut_plot_with_total_summary_and_subgroups_from_dataframe.py datahl.tsv In_Stock Manufacturer --hilolist yes,maybe,no --swap_left_colors
````

That reverses the colors applied the the categories in the left subplot; however, since 'maybe' was in the middle, it remained 'red' with 'no' being blue now. You'll recall, we had hoped to make 'no' red.  
There is another argument that can be added where you can specify how many permutations to step through. While the permutations produced are arbitrary, they are ordered the same each time they are made and so the number of steps to produce a good image can be worked out just once and then the setting can be applied.  
The `--permute_left_colors` followed by an integer can be used like so:

````python
%run donut_plot_with_total_summary_and_subgroups_from_dataframe.py datahl.tsv In_Stock Manufacturer --hilolist yes,maybe,no --permute_left_colors 2
````

In this case the setting `--permute_left_colors 2` seems to be best. Getting the correct setting can take some trial and error; you may wish to set up a loop for choosing setting. Importantly, the number of steps to go that can be supplied starts at one and goes until you exceed the number of possible permutations. (If you try to go too many, you'll be told the number of permutations possible, and so you don't need to figure it out ahead of time. You can supply an outrageous number and get the value possible among the feedback in the error.) Additionally, when more states or subgroups are involved, & hence more permutations, you may find using it along with `--swap_left_colors` may also help.

So now we have a nice plot on the left. Imagine we are very happy with the colors on the left now; however, we aren't happy though with the colors on the right. We started off this section on controlling the color by demonstrating the use of the `--advance_color`.    
There is a similar option, --advance_right_color, that can be used that just applies this advancing of the color palette choices to the right subplot and not both. This also takes an integer indicating the number of steps to take. To demonstate that:

````python
%run donut_plot_with_total_summary_and_subgroups_from_dataframe.py datahl.tsv In_Stock Manufacturer --hilolist yes,maybe,no --permute_left_colors 2 --advance_right_color 20
````

## Changing the plot title

You'll probably want to make the plot titles better reflect your data or delete them. Because sending text from the command line is fraught with issues, I am going to show you ways to specifically edit the script so you can use it form the command line. This will be covered first. 

The situation is easier if you are using importing and calling the main function of the script. That will be covered second.

**Using the script at the command line or equivalent**

I'll provide two ways to do this. The first for advanced command line users and one for those concerned about editing a complex command that has the potential to erase a file.  
If you are comfortable with using complex commands on the command line you could edit the text `NEW TITLE GOES HERE` the following line to change the title.

````python
!sed -i 's/OVERALL/OVERALL TITLE GOES HERE/g' donut_plot_with_total_summary_and_subgroups_from_dataframe.py
!sed -i 's/BY GROUP/NEW TITLE GOES HERE/g' donut_plot_with_total_summary_and_subgroups_from_dataframe.py
````

The changes can be undone with the following:

````python
!sed -i 's/OVERALL TITLE GOES HERE/OVERALL/g' donut_plot_with_total_summary_and_subgroups_from_dataframe.py
!sed -i 's/NEW TITLE GOES HERE/BY GROUP/g' donut_plot_with_total_summary_and_subgroups_from_dataframe.py
````

If you aren't comfortable with editing that complex command, the following makess a function that can be called with the new title.

````python
script_name = "donut_plot_with_total_summary_and_subgroups_from_dataframe.py"
def change_original_titles(o,s):
    '''
    Change the plot titles to the provided text.
    'Overall' plot title is provided first.
    '''
    with open(script_name, 'r') as thefile:
        script=thefile.read()
    script = script.replace('OVERALL', o)
    with open(script_name, 'w') as output_file:
        output_file.write(script)
    with open(script_name, 'r') as thefile:
        script=thefile.read()
    script = script.replace('BY GROUP', s)
    with open(script_name, 'w') as output_file:
        output_file.write(script)
````

Now to use it. Call the function, placing the new title between the quotes, like so:

````python
change_original_titles("OVERALL TITLE GOES HERE","NEW TITLE GOES HERE")
````

You could use the command in the form `change_original_titles('OVERALL TITLE GOES HERE','NEW TITLE GOES HERE')` if you needed to provide double-quotes as part of the title.

To verify that worked:

````python
%run donut_plot_with_total_summary_and_subgroups_from_dataframe.py datao.tsv subgroup group --sort_on_subgroup_name
````

You can provide a space between the quotes or nothing to make no title display.

The problem with this approach is that you can only use those solutions once as they are written.  And the restore commands I provided above only works when changing to the example text, which is probably now what you want to use. You may be better off editing the script directly, or restore the following lines back under 'USER ADJUSTABLE VALUES' in the script if you have to change it more than once from the command line.

```python
total_plot_title = "OVERALL"
group_plot_title = "BY GROUP"
```

Let's set up for the other route by undoing these changes.

````python
!sed -i 's/OVERALL TITLE GOES HERE/OVERALL/g' donut_plot_with_total_summary_and_subgroups_from_dataframe.py
!sed -i 's/NEW TITLE GOES HERE/BY GROUP/g' donut_plot_with_total_summary_and_subgroups_from_dataframe.py
````
